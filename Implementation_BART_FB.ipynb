```python
import numpy as np
import requests
from bs4 import BeautifulSoup
from pygooglenews import GoogleNews
import json
from transformers import pipeline

# Importing all the required libraries like beautiful soup to scrape the webpage, request to request the access to a web page and trasnformers pipeline to summerize
```

```python
url = 'https://www.washingtonpost.com/world/2022/04/29/russia-ukraine-war-news-putin-live-updates/'
num_paras = 10
# Writing the url of the webpage to be scrapped which is the russia-ukraine section war live updates section of the washington post
# Also limiting the maximum number of paragraphs to be scrapped from the page.
news_articles = []
count = 0

# print(count)
# print(str(count) + ". " + entry["link"])
content = requests.get(url).content #The content request to the url to gain permission to scrape.
soup = BeautifulSoup(content, 'html5lib') # Creating the soup variable to do tasks on the available webpage
# print(soup)
news = soup.find_all('div', class_="article-body") #Taking the div section that contains the news updates using BS find_all().
# print(news.get_text())
for i in range(num_paras): #Using for loop to one by one scrape the paragraphs of website and save it to the news_articles variable above
  # print(news[i].get_text())
  text = news[i].get_text()
  # print(type(text))
  if len(text) > 0:
    news_articles.append(text)
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn") #The real summarization ML model 
```

```python
count = 1
for text in news_articles: # Printing the original text
  print(str(count) + ". ", end="")
  print(text)
  count += 1
```

```python
count = 1
for text in news_articles: # Printing the summarized text 
  print(str(count) + ". ", end="")
  print(summarizer(text, max_length=60, min_length=30, do_sample=False)[0]['summary_text'])
  count += 1
```
